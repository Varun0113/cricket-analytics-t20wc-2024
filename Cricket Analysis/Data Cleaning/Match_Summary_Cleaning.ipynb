Data Cleaning 

```python
import pandas as pd

df=pd.DataFrame(pd.read_csv('t20wc.csv'))
df.head()
```

```python
# df.index=df['Match_No']
# del df['Match_No']
df.head()
```

```python
import re
df['Score_Team1'] = df['Score_Team1'].astype(str).str.replace(
    r'\s*\(.*?\)', '', regex=True
)

df['Score_Team2'] = df['Score_Team2'].astype(str).str.replace(
    r'\s*\(.*?\)', '', regex=True
)
df['Ground'] = df['Ground'].astype(str).str.replace(
    r'^.*,\s*', '', regex=True
)
df
```

```python
df.groupby('Ground')['Match_No'].count()
```

```python
df['First_Innings_Runs'] = (
    df['Score_Team1']
    .astype(str)
    .str.extract(r'^(\d+)')   # extract runs before '-'
    .astype(float)
)
df['Second_Innings_Runs'] = (
    df['Score_Team2']
    .astype(str)
    .str.extract(r'^(\d+)')   # extract runs before '-'
    .astype(float)
)
df
```

```python
df.groupby('Ground')['First_Innings_Runs'].mean().round(0)
```

```python
df['Batting_First_Won'] = (
    df['First_Innings_Runs'] > df['Second_Innings_Runs']
)

df['Chasing_Won'] = ~df['Batting_First_Won']
df
```

```python
stadium_analysis = (
    df
    .groupby('Ground', as_index=False)
    .agg(
        Matches=('Match_No', 'count'),
        Avg_First_Innings_Score=('First_Innings_Runs', 'mean'),
        Chasing_Win_Percentage=('Chasing_Won', 'mean')
    )
)

stadium_analysis['Avg_First_Innings_Score'] = stadium_analysis['Avg_First_Innings_Score'].round(1)
stadium_analysis['Chasing_Win_Percentage'] = (stadium_analysis['Chasing_Win_Percentage'] * 100).round(1)

stadium_analysis.sort_values('Matches', ascending=False)
```

```python
df.to_csv('t20wc_clean.csv')
```
