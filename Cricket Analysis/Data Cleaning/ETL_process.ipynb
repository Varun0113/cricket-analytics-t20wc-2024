```python
import pandas as pd
```

```python
data=pd.read_csv('player_info.csv')
df=pd.DataFrame(data)
df
```

```python
df.insert(
    0,
    "player_id",
    ["P" + str(i) for i in range(1, len(df) + 1)]
)
df
```

```python
role_map = {
    "batsman": "bat",
    "bowler": "bowl",
    "batting allrounder": "ar",
    "bowling allrounder": "ar",
    "allrounder": "ar",
    "wicketkeeper": "wk",
    "wicket-keeper": "wk",
    "WK-Batsman": "wk",
    "wk batsman": "wk"
}

df["role_clean"] = (
    df["role"]
    .str.lower()
    .str.replace(r"[-_]", " ", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
    .map(role_map)
)
df.head(10)
```

```python
df["player_name"] = (
    df["player_name"]
    .str.replace(r"\s*(Batsman|Bowler|Batting Allrounder|Bowling Allrounder|Allrounder|Wicketkeeper).*",
                 "",
                 regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

df[df.duplicated(subset=["player_name", "team"], keep=False)]
df
```

```python
df["team"] = (
    df["team"]
    .str.strip()
    .str.title()
)
df
```

```python
bat_map = {
    "right handed bat": "RHB",
    "left handed bat": "LHB"
}

df["batting_style_clean"] = (
    df["batting_style"]
    .str.lower()
    .str.strip()
    .map(bat_map)
)
df
```

```python
df["bowling_arm"] = df["bowling_style"].str.extract(r"(Right|Left)-arm", expand=False)
df["bowling_type"] = df["bowling_style"].str.extract(
    r"(fast-medium|fast|medium|offbreak|legbreak|orthodox)",
    expand=False
)
df
```
